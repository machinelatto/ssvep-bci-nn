```python
from math import log2
import warnings
import pandas as pd
import numpy as np


def itr(n, p, t):
    if p < 0 or 1 < p:
        raise ValueError(
            "stats:itr:BadInputValue " + "Accuracy need to be between 0 and 1."
        )
    elif p < 1 / n:
        itr = 0
    elif p == 1:
        itr = log2(n) * 60 / t
    else:
        itr = (log2(n) + p * log2(p) + (1 - p) * log2((1 - p) / (n - 1))) * 60 / t
    return itr


janelas_segundos = [0.4, 0.6, 0.8, 1]
tempo_desvio_ocular = 0.55
numero_classes = 40
```

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

# Tamanhos das janelas
janelas = [0.4, 0.6, 0.8, 1.0]
x = np.arange(len(janelas))
largura = 0.35

# Load data for paired t-tests (accuracy)
df_proj_04 = pd.read_csv("metricas_treinamento_0.4_segundos.csv")
df_bruto_04 = pd.read_csv("metricas_treinamento_0.4_segundos_dados_brutos.csv")

df_proj_06 = pd.read_csv("metricas_treinamento_0.6_segundos.csv")
df_bruto_06 = pd.read_csv("metricas_treinamento_0.6_segundos_dados_brutos.csv")

df_proj_08 = pd.read_csv("metricas_treinamento_0.8_segundos.csv")
df_bruto_08 = pd.read_csv("metricas_treinamento_0.8_segundos_dados_brutos.csv")

df_proj_1 = pd.read_csv("metricas_treinamento_1_segundos.csv")
df_bruto_1 = pd.read_csv("metricas_treinamento_1_segundos_dados_brutos.csv")

# Get subject means for paired t-tests
acc_proj_04 = df_proj_04.groupby("usuario")["acuracia"].mean().values
acc_bruto_04 = df_bruto_04.groupby("usuario")["acuracia"].mean().values

acc_proj_06 = df_proj_06.groupby("usuario")["acuracia"].mean().values
acc_bruto_06 = df_bruto_06.groupby("usuario")["acuracia"].mean().values

acc_proj_08 = df_proj_08.groupby("usuario")["acuracia"].mean().values
acc_bruto_08 = df_bruto_08.groupby("usuario")["acuracia"].mean().values

acc_proj_1 = df_proj_1.groupby("usuario")["acuracia"].mean().values
acc_bruto_1 = df_bruto_1.groupby("usuario")["acuracia"].mean().values

# Perform paired t-tests
t_stat_04, p_value_04 = stats.ttest_rel(acc_proj_04, acc_bruto_04)
t_stat_06, p_value_06 = stats.ttest_rel(acc_proj_06, acc_bruto_06)
t_stat_08, p_value_08 = stats.ttest_rel(acc_proj_08, acc_bruto_08)
t_stat_1, p_value_1 = stats.ttest_rel(acc_proj_1, acc_bruto_1)

p_values_acc = [p_value_04, p_value_06, p_value_08, p_value_1]


def get_significance_marker(p_value):
    if p_value < 0.001:
        return "***"
    elif p_value < 0.01:
        return "**"
    elif p_value < 0.05:
        return "*"
    else:
        return "ns"


# Acurácia média (convert to percentage)
acuracia_proj = [
    media_geral_proj_04_segundo["acuracia"] * 100,
    media_geral_proj_06_segundo["acuracia"] * 100,
    media_geral_proj_08_segundo["acuracia"] * 100,
    media_geral_proj_1_segundo["acuracia"] * 100,
]

acuracia_bruto = [
    media_geral_bruto_04_segundo["acuracia"] * 100,
    media_geral_bruto_06_segundo["acuracia"] * 100,
    media_geral_bruto_08_segundo["acuracia"] * 100,
    media_geral_bruto_1_segundo["acuracia"] * 100,
]

# Erro padrão (convert to percentage)
erro_proj = [
    erro_padrao_proj_04_segundo["acuracia"] * 100,
    erro_padrao_proj_06_segundo["acuracia"] * 100,
    erro_padrao_proj_08_segundo["acuracia"] * 100,
    erro_padrao_proj_1_segundo["acuracia"] * 100,
]

erro_bruto = [
    erro_padrao_04_segundo["acuracia"] * 100,
    erro_padrao_06_segundo["acuracia"] * 100,
    erro_padrao_08_segundo["acuracia"] * 100,
    erro_padrao_1_segundo["acuracia"] * 100,
]

# Criar gráfico com DPI mais alto para publicação
fig, ax = plt.subplots(figsize=(13, 8), dpi=300)

# Define colors like reference image
color_eegnet = "#1f77b4"  # Blue
color_cca = "#ff7f0e"  # Orange

barras1 = ax.bar(
    x + largura / 2,
    acuracia_proj,
    largura,
    yerr=erro_proj,
    label="EEGNet",
    capsize=8,
    color=color_eegnet,
    error_kw={"linewidth": 2.5, "capthick": 2.5},
)
barras2 = ax.bar(
    x - largura / 2,
    acuracia_bruto,
    largura,
    yerr=erro_bruto,
    label="CCA+EEGNet",
    capsize=8,
    color=color_cca,
    error_kw={"linewidth": 2.5, "capthick": 2.5},
)

# Rótulos e título
ax.set_xlabel("Tamanho do sinal (s)", fontsize=14, fontweight="bold")
ax.set_ylabel("Acurácia (%)", fontsize=14, fontweight="bold")
ax.set_xticks(x)
ax.set_xticklabels([str(j) for j in janelas], fontsize=12)
ax.legend(fontsize=12, loc="upper left", frameon=True, shadow=False)
ax.set_ylim(0, 80)
ax.grid(axis="y", alpha=0.3, linestyle="--", linewidth=1.5)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_linewidth(2)
ax.spines["bottom"].set_linewidth(2)
ax.tick_params(labelsize=11, width=2, length=6)

# Rótulos dentro das barras (branco)
for barras in [barras1, barras2]:
    for bar in barras:
        altura = bar.get_height()
        ax.annotate(
            f"{altura:.2f}",
            xy=(bar.get_x() + bar.get_width() / 2, altura / 2),
            xytext=(0, 0),
            textcoords="offset points",
            ha="center",
            va="center",
            fontsize=11,
            color="white",
            fontweight="bold",
        )

# Add significance markers - positioned above the higher bar
for i in range(len(janelas)):
    max_height = max(acuracia_proj[i] + erro_proj[i], acuracia_bruto[i] + erro_bruto[i])
    marker = get_significance_marker(p_values_acc[i])
    y_pos = max_height + 3
    ax.text(
        x[i], y_pos, marker, ha="center", va="bottom", fontsize=14, fontweight="bold"
    )

plt.tight_layout()
plt.savefig("accuracy_comparison.png", dpi=300, bbox_inches="tight")
plt.show()

# Print p-values
print("Paired t-test p-values (Accuracy):")
for i, p in enumerate(p_values_acc):
    marker = get_significance_marker(p)
    print(f"  {janelas[i]}s: p = {p:.6f} {marker}")
```
